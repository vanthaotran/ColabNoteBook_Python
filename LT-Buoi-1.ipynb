```python
# Viết chương trình nhập hai số A và B, in ra tất cả các số nguyên tố nằm trong
# khoảng [A, B].
a = int(input("a = "))
b = int(input("b = "))
for i in range(a,b+1):
  count = 0
  for j in range(2, i//2+1):
    if i%j==0:
      count = count + 1
  if count == 0:
    print(i)
```

```python
# Viết chương trình nhập vào số n. Kiểm tra xem số n có là số hoàn hảo hay không?
n = int(input("n = "))
sum = 0
for i in range(1, n):
  if n%i==0:
    sum+=i

if(sum == n):
  print("so hoan hao")
else:
  print("khong la so hoan hao")
```
