```python
#  naive_bayes
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
# xay dung du lieu
wine = pd.read_csv("winequality-red (1).csv", sep=";")
x = wine.drop('quality', axis=1)
y = wine["quality"].values
# xay dung mo hinh
model = GaussianNB()
model.fit(x,y)
print(model)
# ung dung mo hinh
kq = model.predict(x)
# hien thi ket qua danh gia
print(kq)
print(metrics.confusion_matrix(y,kq))
print(metrics.classification_report(y,kq))
```
