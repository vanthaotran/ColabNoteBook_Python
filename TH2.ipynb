```python

import numpy as np
import  matplotlib.pyplot as plt

a = np.array(['Java', 'Python', 'PHP', 'Javascript', 'C#', 'C++'])
b = np.array([22,17.5,9,8,7,6])
myExplode = (0.2, 0,0,0,0,0)
plt.bar(a,b)
plt.grid()

plt.pie(b, labels=a,autopct='%1.1f%%', explode=myExplode, shadow=True)
plt.axis('equal') 
plt.show()
```
